#DESAFIO 1

#CENÁRIO
Uma empresa de mídia digital, chamada "NewsStream", está desenvolvendo um sistema de recomendação de artigos para melhorar a experiência de usuário em seu site.
Para fornecer recomendações mais personalizadas e contextuais, a empresa quer preencher automaticamente lacunas em resumos e artigos com base em palavras-chave relevantes.
Isso ajudará a fornecer sugestões mais precisas e envolventes aos leitores, especialmente para artigos que abordam tópicos populares ou em alta.

#PROBLEMA
NewsStream quer preencher automaticamente a seguinte frase de resumo de artigo com uma palavra que seja mais relevante possível:
"Artificial Intelligence has been revolutionizing various industries, including healthcare, finance and [mask]."

```python
#Solução Proposta de Resolução do Desafio-1  - Modelo Padrão
from transformers import pipeline
model = pipeline('fill-mask')
phrase = 'Artificial Intelligence has been revolutionizing various industries, including healthcare, finance and <mask>.'
predictions = model(phrase)
print('='*165)
print('NewsStream - Sistema de Recomendação'.center(165))
print('='*165)
for predict in predictions:
  score = predict['score']
  label = predict['token_str']
  print(f'Sugestões de Frase: Artificial Intelligence has been revolutionizing various industries, including healthcare, finance and {label} | Score: {score*100:.2f}%')
```

```python
#Solução Proposta de Resolução do Desafio-2  - Modelo Brasileiro
from transformers import pipeline

#definição de modelo e tarefa
modelo = 'rufimelo/Legal-BERTimbau-large'
tarefa = 'fill-mask'
model = pipeline(task=tarefa, model=modelo)
frase = 'O advogado apresentou [MASK] ao juíz.'
predictions = model(frase)
print('='*165)
print('PREVISÕES COM MODELO {modelo.upper()}'.center(165))
print('='*165)
for predict in predictions:
  score = predict['score']
  label = predict['token_str']
  print(f'Sugestões de Frase: O advogado apresentou {label} ao juíz. | Score: {score*100:.2f}%')
```

#Classificação de Textos

```python
from transformers import pipeline

#Carregando pipeline de classificação de tópicos
classifier = pipeline('text-classification', model='distilbert-base-uncased')

#Texto para rotulagen de negócios
texts = [
    'The company reported a 10% incrase in revenue this quarter.',
    'Our marketing team launched a sucessful social media campaign.',
    'We are investing in cutting-edge AI technology to improve our services'
]

#Realizar a classificação de texto
results = classifier(texts)

#Exibir o resultado com rótulos customizados
labels = ['Business', 'Marketing', 'Technology']
for text, result in zip (texts, results):
  #Simular atribuição de rótulos com base no conteúdo
  if 'revenue' in text or 'stock' in text:
    label = 'Business'
  elif 'Marketing' in text or 'Campaign' in text:
    label = 'Marketing'
  elif 'Technology' in text or 'AI' in text:
    label = 'Technology'
  else:
    label = 'Other'

  print(f"Texto: {text}\nRótulo: {label}, Probabilidade: {result['score']:.4f}\n")
```

#DESAFIO 2

#CENÁRIO
A empresa "SecureMail" é um provedor de serviços de e-mail que atende a milhares de usuários em todo o mundo.
Recentemente, houve um aumento significativo na quantidade de e-mails de spam recebidos por seus usuários, o que resultou em várias reclamações e uma possível perda de confiança no serviço.
Para resolver esse problema, a equipe de desenvolvimento SecureMail, decide implementar uma solução de filtragem de spam baseada em modelos de aprendizado de máquina.

#PROBLEMA
A SecureMail precisa desenvolver um sistema eficiente de filtragem de spam que possa analisar os e-mails recebidos e classificá-los como "Spam" ou "Não Spam" com alta precisão.
A empresa decide realizar um pipeline do Tranformers da Hugging Face para esta tarefa, aproveitando modelos pré-treinados para classificação de texto.

```python
#Solução Desafio 2
from transformers import pipeline

#Carregando pipeline de classificação de texto
spam_classifier = pipeline('text-classification', model='distilbert-base-uncased-finetuned-sst-2-english')

#e-mails para classificação
emails = [
    'Congratulations! You,ve won a $1000 gift card. Click here to claim your prize.',
    'Dont miss out on our excelusive offer. Buy now and save 50%.',
    'Dear customer, your recent order has been shipped and is on its way.',
    'Important update: Your account needs verification. Please login to secure your informations'
]
#Realizar a classificação de spam
results = spam_classifier(emails)

#Exibir os resultados
for email, result in zip(emails, results):
  label = 'Não Spam' if result["label"] == "LABEL_1" else "Spam"
  print(f'E-mail: {email}\nRótulo: {label}, Probabilidade: {result["score"]:.4f}\n')
```

#ZERO-SHOT CLASSIFICATION
É uma técnica em aprendizado de máquina onde um modelo é capaz de classificar dados que não foram vistos durante o treinamento.

Ex: Rótulos não vistos podem ser determinados, Sem treinamento específico, Útil quando há falta de recursos para treinar um novo modelo.

#CÓDIGO_4 USANDO O HUGGING FACE (ZERO_SHOT CLASSIFICATION)

```python
from transformers import pipeline

#Carregando pipeline de zero-shot classification
classifier = pipeline('zero-shot-classification')

#Texto a ser classificado
text = 'The company is launching a new AI product next month.'

#Candidatos a rótulos (Classes)
candidate_labels = ['Technology', 'Business', 'Sports']

#Realizar a classificação zero-shot
result = classifier(text, candidate_labels)

#Exibir o resultado
print(result)
```

#CÓDIGO_5 USANDO O HUGGING FACE (SUMMARIZATION)
A tarefa de summarization (resumo) no pipeline do Transformers da Hugging Face, é usada para gerar um resumo conciso e significativo de um texto mais longo.

```python
from transformers import pipeline
summarizer = pipeline('summarization', model='cnmoro/ptt5-base-ptbr-summarization')

#Texto em portugues para ser resumido
text = """
A inteligência artificial (IA) refere-se à simulação de processos de inteligência humana por meio de máquinas , especialmente sistemas de computador.
Esses processos incluem aprendizado (a aquisição de informações e regras para o uso das informações), raciocínio (usando regras para alcançar conclusões aproximadas ou definitivas).
Aplicações específicas de IA incluem sistemas especialistas, reconhecimento de voz e visão, e processamento de linguagem natural.
A IA tem o potencial de tranformar diversos setores, incluindo a saúde, finanças, e transporte, fornecendo novas maneiras de resolver problemas complexos e melhorar a eficiência.
"""

#Realizar resumo do texto
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)

#Exibir o resumo
print(f'Resumo: {summary[0]["summary_text"]}')
```
